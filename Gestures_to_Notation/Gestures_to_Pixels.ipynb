```python
from PIL import Image
import numpy as np
```

```python
im = Image.open('pix.png')
print(im.mode)
pix_val = list(im.getdata())
pix_val_flat = [x for sets in pix_val for x in sets]

im = im.convert(mode="1", dither=1)
```

```python
pix_val = list(im.getdata())
pix_mat = np.asmatrix(pix_val)
pix_mat = pix_mat.reshape(16,16)
pix_mat = np.flip(pix_mat, 0)
pix_mat = pix_mat.getT()
pix_mat = np.asarray(pix_mat)
```

```python
##expecting [0,1,1,3]

final_arr = []
for i in range(len(pix_mat)):
    for j in range(len(pix_mat[i])):
        if pix_mat[i][j] == 0:
            final_arr.append(j)
#print(final_arr)
```

```python
#write to CSV
import pandas as pd 
pd.DataFrame(final_arr).to_csv("pix.csv")
```
